We aim to gather data via energy.gov on the current locations of the charging stations in the area for model evaluation and improvement. We will get the data by
requesting the API url with our own developer API key and the queried parameters related to electric vehicle stations.

```python
import requests
import pandas as pd
import xml.etree.ElementTree as ET
from io import StringIO

# Define the URL
url = "https://developer.nrel.gov/api/alt-fuel-stations/v1.csv"

# Define query parameters
query_params = {
    "api_key": "9tTOLG97gY6odZamjcPb1ljjFjNbDjVoeaR7h3yQ",
    "state": "WA",
    "access" : "public",
    "fuel_type" : "ELEC",
    'ev_charging_level': '2'
}

# Make the GET request
response = requests.get(url, params=query_params)
```

```python
df = pd.read_csv(StringIO(response.text))
df = df.loc[df["City"] == "Seattle"]
df = df[['Station Name', 'City', 'State', 'ZIP', 'Status Code', 'Expected Date', 'Owner Type Code',
        'Geocode Status', 'Latitude', 'Longitude', 'Facility Type', 'Restricted Access']]
df = df.loc[df["Restricted Access"] != True]
df
```
